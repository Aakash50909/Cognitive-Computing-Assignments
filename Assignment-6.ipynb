```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Q1
M = float(input("Enter a value for M: "))
x = np.linspace(-10, 10, 400)
y1 = M * x**2
y2 = M * np.sin(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y1, 'r--', label='y = M * x^2')
plt.plot(x, y2, 'b-', label='y = M * sin(x)')
plt.title("Mathematical Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

# Q2
subjects = ['Math', 'Science', 'English', 'History', 'Art']
scores = [85, 90, 78, 88, 92]
df = pd.DataFrame({'Subject': subjects, 'Score': scores})
plt.figure(figsize=(8, 5))
barplot = sns.barplot(x='Subject', y='Score', data=df, palette='muted')
for i in range(len(scores)):
    barplot.text(i, scores[i] + 1, str(scores[i]), ha='center')
plt.title("Subject Scores")
plt.xlabel("Subject")
plt.ylabel("Score")
plt.grid(axis='y')
plt.show()

# Q3
roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)
data = np.random.randn(50)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes[0, 0].plot(np.cumsum(data))
axes[0, 0].set_title("Cumulative Sum")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumsum")
axes[0, 0].grid(True)
axes[0, 1].scatter(range(50), data)
axes[0, 1].set_title("Random Noise Scatter")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Value")
axes[0, 1].grid(True)
fig.delaxes(axes[1, 0])
fig.delaxes(axes[1, 1])
plt.tight_layout()
plt.show()

# Q4
url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
data = pd.read_csv(url)
plt.figure(figsize=(10, 6))
plt.plot(data['month_number'], data['total_profit'], marker='o')
plt.title("Total Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()
plt.figure(figsize=(12, 8))
for column in data.columns[1:-1]:
    plt.plot(data['month_number'], data[column], label=column)
plt.title("Product Sales Data")
plt.xlabel("Month")
plt.ylabel("Sales Units")
plt.legend()
plt.grid(True)
plt.show()
plt.figure(figsize=(10, 6))
data.iloc[:, 1:].sum().plot(kind='bar', color=sns.color_palette('hls', len(data.columns[1:])))
plt.title("Total Sales of All Products")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.grid(axis='y')
plt.show()
```

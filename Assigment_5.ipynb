```python
#Question 3
#ASCII SUM = A+C = 65+67= 132
X=132
sales=np.array([132,182,232,282,332])
def personalised_tax_rate(x):
  return (x-((x%5)+5)/100)
print(np.vectorize(personalised_tax_rate)(sales))
#Adjust sales based on discount:
# If sales < X+100, apply a 5% discount.
#  If sales >= X+100, apply a 10% discount.
def discount(Z):
  if(Z<X+100):
    return (Z-(0.05*Z))
  else:
    return (Z-(0.1*Z))
print(np.vectorize(discount)(sales))
sales_matrix=np.tile(sales, (3, 1))
weeks=np.arange(3).reshape(3, 1)
adjusted_sales=sales_matrix * (1 + 0.02 * weeks)
print("Original Sales Matrix:\n",sales_matrix)
print("Adjusted Sales Matrix (with 2% weekly increase):\n",adjusted_sales)
```

```python
# Question 4
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label=r"$y = x^2$", linestyle='dashed', color='blue')
plt.plot(x, y2, label=r"$y = \sin(x)$", linestyle='solid', color='green')
plt.plot(x, y3, label=r"$y = e^x$", linestyle='dotted', color='red')
plt.plot(x, y4, label=r"$y = \log(|x| + 1)$", linestyle='dashdot', color='purple')
plt.title("Plots of Various Functions", fontsize=14)
plt.xlabel("x values", fontsize=12)
plt.ylabel("y values", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()

```
